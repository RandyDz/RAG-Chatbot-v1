````python
import os
import pandas as pd
from pathlib import Path
from unstructured.partition.pdf import partition_pdf
from unstructured.partition.docx import partition_docx
from unstructured.partition.pptx import partition_pptx
import logging


SUMBER_FOLDER = Path("/content/drive/MyDrive/data_sample")

MARKDOWN_FOLDER = Path("/content/drive/MyDrive/data_sample/markdown_file")


logging.getLogger("unstructured").setLevel(logging.ERROR)
logging.getLogger("unstructured_inference").setLevel(logging.ERROR)

# Konversi file

def konversi_pdf(file_path, output_path):
    """Mengekstrak teks dari PDF dan menyimpannya sebagai Markdown."""
    print(f"  Mengonversi PDF: {file_path.name}...")
    try:
        elements = partition_pdf(filename=str(file_path), strategy="fast")
        with open(output_path, "w", encoding="utf-8") as f:
            for el in elements:
                # Mengonversi elemen (misal: Title, Text) ke string
                f.write(str(el) + "\n\n")
    except Exception as e:
        print(f"    GAGAL memproses {file_path.name}: {e}")

def konversi_docx(file_path, output_path):
    """Mengekstrak teks dari DOCX dan menyimpannya sebagai Markdown."""
    print(f"  Mengonversi DOCX: {file_path.name}...")
    try:
        elements = partition_docx(filename=str(file_path))
        with open(output_path, "w", encoding="utf-8") as f:
            for el in elements:
                f.write(str(el) + "\n\n")
    except Exception as e:
        print(f"    GAGAL memproses {file_path.name}: {e}")

def konversi_pptx(file_path, output_path):
    """Mengekstrak teks dari PPTX dan menyimpannya sebagai Markdown."""
    print(f"  Mengonversi PPTX: {file_path.name}...")
    try:
        elements = partition_pptx(filename=str(file_path))
        with open(output_path, "w", encoding="utf-8") as f:
            for el in elements:
                f.write(str(el) + "\n\n")
    except Exception as e:
        print(f"    GAGAL memproses {file_path.name}: {e}")

def konversi_excel(file_path, output_path):
    """Mengonversi SETIAP SHEET di Excel menjadi tabel Markdown."""
    print(f"  Mengonversi XLSX: {file_path.name}...")
    try:
        xls = pd.ExcelFile(file_path)
        with open(output_path, "w", encoding="utf-8") as f:
            for sheet_name in xls.sheet_names:
                df = pd.read_excel(xls, sheet_name=sheet_name)

                # Menulis nama sheet sebagai judul
                f.write(f"## Sheet: {sheet_name}\n\n")

                # Mengonversi DataFrame ke tabel Markdown
                markdown_table = df.to_markdown(index=False)
                f.write(markdown_table + "\n\n")
    except Exception as e:
        print(f"    GAGAL memproses {file_path.name}: {e}")

def konversi_txt(file_path, output_path):
    """Menyalin konten TXT ke file .md baru."""
    print(f"Mengonversi TXT: {file_path.name}...")
    try:
        with open(file_path, "r", encoding="utf-8") as f_in:
            content = f_in.read()
        with open(output_path, "w", encoding="utf-8") as f_out:
            f_out.write(content)
    except Exception as e:
        print(f"GAGAL memproses {file_path.name}: {e}")


def main():
    print(f"Memulai proses konversi di folder: {SUMBER_FOLDER}")


    MARKDOWN_FOLDER.mkdir(exist_ok=True)

    print("Memindai file...")
    list_semua_file = list(SUMBER_FOLDER.rglob("*"))

    if not list_semua_file:
        print(f"Tidak ada file yang ditemukan di {SUMBER_FOLDER}. Pastikan Anda meletakkan file di sana.")
        return

    print(f"Total file ditemukan: {len(list_semua_file)}")

    # konversi file satu per satu
    for file_path in list_semua_file:

        # Melewatkan folder, hanya proses file
        if not file_path.is_file():
            continue

        # Dapatkan ekstensi file (.pdf, .docx,...)
        suffix = file_path.suffix.lower()

        #.pdf to .md
        output_file_name = file_path.stem + ".md"
        output_path = MARKDOWN_FOLDER / output_file_name


        if suffix == ".pdf":
            konversi_pdf(file_path, output_path)
        elif suffix == ".docx":
            konversi_docx(file_path, output_path)
        elif suffix == ".pptx":
            konversi_pptx(file_path, output_path)
        elif suffix == ".xlsx":
            konversi_excel(file_path, output_path)
        elif suffix == ".txt":
            konversi_txt(file_path, output_path)
        else:
            print(f"  Melewatkan file (format tidak didukung): {file_path.name}")

    print("-" * 30)
    print(f"Proses Selesai. File Markdown disimpan di: {MARKDOWN_FOLDER}")

if __name__ == "__main__":
    main()
````

````python
from notion_client import Client
import os

MARKDOWN_FOLDER = "/content/drive/MyDrive/data_sample/markdown_file"

LIST_ID_SUMBER = [
    "18477ecfd6c9801f84f1d55cd4506bfc",
    "17677ecfd6c98063a562d3b7f10300de",
    "17377ecfd6c98079b53eea61e5de0c42"
]

# Initialize Notion client
notion = Client(auth="INSERT_NOTION_API_KEY_HERE")

def block_to_markdown(block):
    """Convert Notion block to Markdown string"""
    block_type = block.get('type')

    # Paragraph
    if block_type == 'paragraph':
        return rich_text_to_markdown(block['paragraph'].get('rich_text', [])) + '\n\n'

    # Headings
    elif block_type == 'heading_1':
        return '# ' + rich_text_to_markdown(block['heading_1'].get('rich_text', [])) + '\n\n'
    elif block_type == 'heading_2':
        return '## ' + rich_text_to_markdown(block['heading_2'].get('rich_text', [])) + '\n\n'
    elif block_type == 'heading_3':
        return '### ' + rich_text_to_markdown(block['heading_3'].get('rich_text', [])) + '\n\n'

    # Lists
    elif block_type == 'bulleted_list_item':
        return '- ' + rich_text_to_markdown(block['bulleted_list_item'].get('rich_text', [])) + '\n'
    elif block_type == 'numbered_list_item':
        return '1. ' + rich_text_to_markdown(block['numbered_list_item'].get('rich_text', [])) + '\n'

    # Code
    elif block_type == 'code':
        language = block['code'].get('language', '')
        code_text = rich_text_to_markdown(block['code'].get('rich_text', []))
        return f'```{language}\n{code_text}\n```\n\n'

    # Quote
    elif block_type == 'quote':
        return '> ' + rich_text_to_markdown(block['quote'].get('rich_text', [])) + '\n\n'

    # Divider
    elif block_type == 'divider':
        return '---\n\n'

    # To-do
    elif block_type == 'to_do':
        checked = block['to_do'].get('checked', False)
        checkbox = '[x]' if checked else '[ ]'
        return f'{checkbox} ' + rich_text_to_markdown(block['to_do'].get('rich_text', [])) + '\n'

    # Toggle
    elif block_type == 'toggle':
        return rich_text_to_markdown(block['toggle'].get('rich_text', [])) + '\n\n'

    # Callout
    elif block_type == 'callout':
        return '> 💡 ' + rich_text_to_markdown(block['callout'].get('rich_text', [])) + '\n\n'

    # Default
    return ''

def rich_text_to_markdown(rich_text_array):
    """Convert Notion rich text to Markdown"""
    result = ''
    for text_obj in rich_text_array:
        content = text_obj.get('plain_text', '')
        annotations = text_obj.get('annotations', {})

        # Apply formatting
        if annotations.get('bold'):
            content = f'**{content}**'
        if annotations.get('italic'):
            content = f'*{content}*'
        if annotations.get('strikethrough'):
            content = f'~~{content}~~'
        if annotations.get('code'):
            content = f'`{content}`'

        # Handle links
        if text_obj.get('href'):
            content = f'[{content}]({text_obj["href"]})'

        result += content

    return result

def get_page_title(page_id):
    """Get page title"""
    try:
        page = notion.pages.retrieve(page_id)
        properties = page.get('properties', {})

        for prop_name, prop_value in properties.items():
            if prop_value.get('type') == 'title':
                title_content = prop_value.get('title', [])
                if title_content:
                    return title_content[0].get('plain_text', page_id)
        return page_id
    except:
        return page_id

def get_all_blocks(block_id):
    """Recursively get all blocks including children"""
    blocks = []

    try:
        # Get blocks with pagination
        has_more = True
        start_cursor = None

        while has_more:
            if start_cursor:
                response = notion.blocks.children.list(
                    block_id=block_id,
                    start_cursor=start_cursor
                )
            else:
                response = notion.blocks.children.list(block_id=block_id)

            blocks.extend(response.get('results', []))
            has_more = response.get('has_more', False)
            start_cursor = response.get('next_cursor')

        # Get children blocks recursively
        for block in blocks:
            if block.get('has_children'):
                child_blocks = get_all_blocks(block['id'])
                block['children'] = child_blocks

        return blocks
    except Exception as e:
        print(f"Error getting blocks: {e}")
        return []

def convert_page_to_markdown(page_id):
    """Convert entire Notion page to Markdown"""
    try:
        # Clean page ID
        clean_id = page_id.replace('-', '')

        # Get page title
        title = get_page_title(clean_id)

        # Get all blocks
        blocks = get_all_blocks(clean_id)

        # Convert to markdown
        markdown_content = f"# {title}\n\n"

        for block in blocks:
            markdown_content += block_to_markdown(block)

            # Handle children blocks (indentation)
            if block.get('children'):
                for child in block['children']:
                    child_md = block_to_markdown(child)
                    # Indent children
                    markdown_content += '  ' + child_md.replace('\n', '\n  ')

        return title, markdown_content
    except Exception as e:
        print(f"Error converting page: {e}")
        return None, None

# Main execution
os.makedirs(MARKDOWN_FOLDER, exist_ok=True)

print("="*60)
print("MEMULAI KONVERSI NOTION KE MARKDOWN")
print("="*60)

successful = 0
failed = 0

for page_id in LIST_ID_SUMBER:
    print(f"\nMemproses: {page_id}")

    title, markdown = convert_page_to_markdown(page_id)

    if markdown:
        # Sanitize filename
        safe_title = "".join(c for c in title if c.isalnum() or c in (' ', '-', '_')).strip()
        safe_title = safe_title[:100] or page_id

        # Save to file
        output_path = os.path.join(MARKDOWN_FOLDER, f"{safe_title}.md")

        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(markdown)

        print(f"✓ BERHASIL: {safe_title}.md")
        successful += 1
    else:
        print(f"✗ GAGAL: {page_id}")
        failed += 1

print("\n" + "="*60)
print("RINGKASAN")
print("="*60)
print(f"Berhasil: {successful}")
print(f"Gagal: {failed}")
print(f"File tersimpan di: {MARKDOWN_FOLDER}")
````

````python
import re
import os

def preprocess_text(text:str) -> str:
  # Hapus multiple spaces jadi single space
  text = re.sub(r' {2,}', ' ', text)

  # Hapus multiple newlines (lebih dari 2) jadi 2 newlines
  text = re.sub(r'\n{3,}', '\n\n', text)

  # Hapus trailing whitespace di setiap baris
  lines = text.split('\n')
  text = '\n'.join(line.rstrip() for line in lines)

  # Hapus HTML comments: <!-- ... -->
  text = re.sub(r'<!--.*?-->', '', text, flags=re.DOTALL)

  # Hapus HTML tags: <tag>...</tag>
  text = re.sub(r'<[^>]+>', '', text)

  # Hapus empty markdown links []()
  text = re.sub(r'\[\]\(\)', '', text)

  # Hapus links dengan text kosong [](url)
  text = re.sub(r'\[\]\([^)]+\)', '', text)

  # Hapus baris yang hanya berisi angka (page numbers)
  text = re.sub(r'^\s*\d+\s*$', '', text, flags=re.MULTILINE)

  text = text.encode('utf-8', 'ignore').decode('utf-8') #Encode dan decode utk hapus karakter yg corrupt

  # Hapus dashes atau underscores yang panjang (separator lines)
  text = re.sub(r'[-_]{4,}', '', text)

  # Hapus bullet points yang jadi karakter aneh dari konversi
  text = re.sub(r'[\u2022\u25CF\u25CB\u25A0\u25A1\u25AA\u25AB]', '-', text)
  return text

def clean_markdown_folder(input_folder: str, output_folder: str = None):
    """
    Bersihkan semua file markdown dalam folder

    Args:
        input_folder: Folder berisi file markdown
        output_folder: Folder untuk menyimpan hasil (optional)
    """
    print(f"\n{'='*70}")
    print(f"BATCH CLEANING - Folder: {input_folder}")
    print(f"{'='*70}")

    # Setup output folder
    if output_folder is None:
        output_folder = input_folder + "_cleaned"

    os.makedirs(output_folder, exist_ok=True)

    # Cari semua file .md
    md_files = [f for f in os.listdir(input_folder) if f.endswith('.md')]

    print(f"\nDitemukan {len(md_files)} file markdown")

    results = {
        'successful': [],
        'failed': []
    }

    # Process setiap file
    for i, filename in enumerate(md_files, 1):
        print(f"\n[{i}/{len(md_files)}] Processing: {filename}")

        try:
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)

            # Baca
            with open(input_path, 'r', encoding='utf-8') as f:
                text = f.read()

            # Clean
            cleaned_text = preprocess_text(text)

            # Simpan
            with open(output_path, 'w', encoding='utf-8') as f:
                f.write(cleaned_text)

            results['successful'].append(filename)

        except Exception as e:
            print(f"  Error: {e}")
            results['failed'].append(filename)

    # Summary
    print(f"\n{'='*70}")
    print("SUMMARY")
    print(f"{'='*70}")
    print(f"Berhasil: {len(results['successful'])} files")
    print(f"Gagal:    {len(results['failed'])} files")
    print(f"Output folder: {output_folder}")

    return results

cleaned = clean_markdown_folder(
input_folder="/content/drive/MyDrive/data_sample/markdown_file",
output_folder="/content/drive/MyDrive/data_sample/markdown_file")
````

````python
import os
import re
from typing import List, Dict, Any
from datetime import datetime

try:
    import tiktoken
    TIKTOKEN_AVAILABLE = True
except ImportError:
    TIKTOKEN_AVAILABLE = False
    print("⚠️  tiktoken tidak terinstall. Install dengan: pip install tiktoken")

class RecursiveChunker:
    """
    Memecah teks panjang menjadi chunks dengan recursive splitting
    """

    def __init__(self, chunk_size=1000, chunk_overlap=200):
        """
        Args:
            chunk_size: Target size per chunk (dalam characters)
            chunk_overlap: Overlap antar chunks untuk maintain context
        """
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap

        # Setup token counter jika tersedia
        if TIKTOKEN_AVAILABLE:
            self.encoding = tiktoken.get_encoding("cl100k_base")

        # Separator hierarchy - dari paling besar ke paling kecil
        self.separators = [
            "\n\n\n",   # Multiple newlines (section breaks)
            "\n\n",     # Paragraph breaks
            "\n",       # Line breaks
            ". ",       # Sentences
            "! ",       # Exclamation
            "? ",       # Question
            "; ",       # Semicolon
            ", ",       # Comma
            " ",        # Space (words)
            ""          # Characters (last resort)
        ]

    def count_length(self, text: str) -> int:
        """
        Hitung panjang text (characters atau tokens jika tersedia)
        """
        if TIKTOKEN_AVAILABLE:
            return len(self.encoding.encode(text))
        else:
            return len(text)

    def split_text(self, text: str, separators: List[str]) -> List[str]:
        """
        Split text menggunakan separator pertama yang ditemukan
        """
        # Coba separator pertama
        separator = separators[0] if separators else ""

        if separator:
            splits = text.split(separator)
        else:
            # Kalau sudah tidak ada separator, split per character
            splits = list(text)

        # Filter out empty strings
        splits = [s for s in splits if s.strip()]

        return splits, separator

    def merge_splits(self, splits: List[str], separator: str) -> List[str]:
        """
        Merge splits menjadi chunks dengan size yang sesuai
        """
        chunks = []
        current_chunk = []
        current_length = 0

        for split in splits:
            split_length = self.count_length(split)

            # Cek apakah menambahkan split ini akan exceed chunk_size
            if current_length + split_length > self.chunk_size and current_chunk:
                # Simpan chunk saat ini
                chunk_text = separator.join(current_chunk)
                chunks.append(chunk_text)

                # Mulai chunk baru dengan overlap
                # Ambil beberapa split terakhir untuk overlap
                overlap_splits = []
                overlap_length = 0

                for prev_split in reversed(current_chunk):
                    prev_length = self.count_length(prev_split)
                    if overlap_length + prev_length <= self.chunk_overlap:
                        overlap_splits.insert(0, prev_split)
                        overlap_length += prev_length
                    else:
                        break

                current_chunk = overlap_splits
                current_length = overlap_length

            # Tambahkan split ke current chunk
            current_chunk.append(split)
            current_length += split_length

        # Jangan lupa chunk terakhir
        if current_chunk:
            chunk_text = separator.join(current_chunk)
            chunks.append(chunk_text)

        return chunks

    def recursive_split(self, text: str, separators: List[str]) -> List[str]:
        """
        Recursive splitting - coba separator dari yang terbesar
        """
        # Base case: text sudah cukup kecil
        if self.count_length(text) <= self.chunk_size:
            return [text]

        # Kalau tidak ada separator lagi, return text as is
        if not separators:
            return [text]

        # Split dengan separator pertama
        splits, separator = self.split_text(text, separators)

        # Merge splits menjadi chunks
        chunks = self.merge_splits(splits, separator)

        # Recursive: untuk setiap chunk yang masih terlalu besar,
        # split lagi dengan separator berikutnya
        final_chunks = []
        for chunk in chunks:
            if self.count_length(chunk) > self.chunk_size:
                # Chunk masih terlalu besar, split lagi dengan separator berikutnya
                sub_chunks = self.recursive_split(chunk, separators[1:])
                final_chunks.extend(sub_chunks)
            else:
                final_chunks.append(chunk)

        return final_chunks

    def chunk(self, text: str) -> List[str]:
        """
        Main chunking function
        """
        print(f"\n📦 MEMULAI CHUNKING...")
        print(f"  Chunk size: {self.chunk_size} {'tokens' if TIKTOKEN_AVAILABLE else 'chars'}")
        print(f"  Overlap: {self.chunk_overlap} {'tokens' if TIKTOKEN_AVAILABLE else 'chars'}")

        # Lakukan recursive splitting
        chunks = self.recursive_split(text, self.separators)

        # Filter chunks yang terlalu kecil (< 10 chars)
        chunks = [c for c in chunks if len(c.strip()) > 10]

        # Statistik
        total_length = self.count_length(text)
        avg_chunk_length = sum(self.count_length(c) for c in chunks) / len(chunks) if chunks else 0

        print(f"\n✓ Chunking selesai!")
        print(f"  Total chunks: {len(chunks)}")
        print(f"  Original length: {total_length:,} {'tokens' if TIKTOKEN_AVAILABLE else 'chars'}")
        print(f"  Avg chunk length: {avg_chunk_length:.0f} {'tokens' if TIKTOKEN_AVAILABLE else 'chars'}")

        return chunks

````

````python
class MetadataExtractor:

    @staticmethod
    def extract_title(text: str) -> str:

        # Cari heading level 1
        match = re.search(r'^#\s+(.+)$', text, re.MULTILINE)
        if match:
            return match.group(1).strip()

        # jika tidak exist ambil baris pertama
        first_line = text.split('\n')[0].strip()
        return first_line[:100] if first_line else "Untitled"

    @staticmethod
    def extract_headings(text: str) -> List[str]:
        #Extract all headings
        headings = re.findall(r'^#{1,6}\s+(.+)$', text, re.MULTILINE)
        return headings

    @staticmethod
    def create_metadata(text: str, file_path: str) -> Dict[str, Any]:
        #metadata dict
        return {
            'source_file': os.path.basename(file_path),
            'file_path': file_path,
            'title': MetadataExtractor.extract_title(text),
            'headings': MetadataExtractor.extract_headings(text)[:10],
            'char_count': len(text),
            'word_count': len(text.split()),
            'processed_at': datetime.now().isoformat()
        }

````

````python
class ChunkEnricher:
    """
    Menambahkan metadata dan context ke setiap chunk
    """

    @staticmethod
    def enrich_chunks(chunks: List[str], metadata: Dict[str, Any]) -> List[Dict[str, Any]]:
        """
        Tambahkan metadata dan context ke setiap chunk
        """
        print(f"\n✨ ENRICHING CHUNKS dengan metadata...")

        enriched_chunks = []

        for i, chunk_text in enumerate(chunks):
            # Hitung tokens jika tersedia
            if TIKTOKEN_AVAILABLE:
                encoding = tiktoken.get_encoding("cl100k_base")
                token_count = len(encoding.encode(chunk_text))
            else:
                token_count = len(chunk_text)

            chunk_data = {
                # Chunk identification
                'chunk_id': f"{metadata['source_file']}_{i}",
                'chunk_index': i,
                'total_chunks': len(chunks),

                # Chunk content
                'text': chunk_text,
                'char_count': len(chunk_text),
                'word_count': len(chunk_text.split()),
                'token_count': token_count,

                # Context - untuk maintain coherence
                'previous_text': chunks[i-1][-200:] if i > 0 else None,
                'next_text': chunks[i+1][:200] if i < len(chunks) - 1 else None,

                # Document metadata
                'source_file': metadata['source_file'],
                'file_path': metadata['file_path'],
                'document_title': metadata['title'],
                'processed_at': metadata['processed_at']
            }

            enriched_chunks.append(chunk_data)

        print(f"  ✓ {len(enriched_chunks)} chunks diperkaya dengan metadata")

        return enriched_chunks
````

````python
def chunk_single_file(file_path: str, chunk_size=1000, chunk_overlap=200):

    print(f"\n{'='*70}")
    print(f"CHUNKING FILE: {os.path.basename(file_path)}")
    print(f"{'='*70}")

    # read file
    print("\n📖 Membaca file...")
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()
    print(f"  File size: {len(text):,} characters")

    # Extract metadata
    print("\n📋 Extracting metadata...")
    metadata = MetadataExtractor.create_metadata(text, file_path)
    print(f"  Title: {metadata['title']}")
    print(f"  Headings: {len(metadata['headings'])} found")

    # Chunking
    chunker = RecursiveChunker(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    chunks = chunker.chunk(text)

    # Enrich chunks
    enriched_chunks = ChunkEnricher.enrich_chunks(chunks, metadata)

    return enriched_chunks

def chunk_folder(input_folder: str, output_folder: str = None,
                 chunk_size=1000, chunk_overlap=200):

    print(f"\n{'='*70}")
    print(f"BATCH CHUNKING - Folder: {input_folder}")
    print(f"{'='*70}")

    # Cari semua .md files
    md_files = [f for f in os.listdir(input_folder) if f.endswith('.md')]
    print(f"\n📁 Ditemukan {len(md_files)} markdown files")

    all_chunks = {}
    total_chunks = 0

    # Process setiap file
    for i, filename in enumerate(md_files, 1):
        print(f"\n{'─'*70}")
        print(f"[{i}/{len(md_files)}] Processing: {filename}")
        print(f"{'─'*70}")

        try:
            file_path = os.path.join(input_folder, filename)
            chunks = chunk_single_file(file_path, chunk_size, chunk_overlap)

            all_chunks[filename] = chunks
            total_chunks += len(chunks)

        except Exception as e:
            print(f"✗ Error processing {filename}: {e}")

    # Summary
    print(f"\n{'='*70}")
    print("CHUNKING SUMMARY")
    print(f"{'='*70}")
    print(f"Files processed: {len(all_chunks)}")
    print(f"Total chunks created: {total_chunks}")
    print(f"Average chunks per file: {total_chunks/len(all_chunks):.1f}")


    if output_folder:
        import json
        os.makedirs(output_folder, exist_ok=True)

        output_file = os.path.join(output_folder, 'chunks_metadata.json')
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(all_chunks, f, indent=2, ensure_ascii=False)

        print(f"\n💾 Chunks metadata saved to: {output_file}")

    return all_chunks

def display_chunk_preview(chunks: List[Dict], num_preview=3):
    """
    Tampilkan preview beberapa chunks pertama
    """
    print(f"\n{'='*70}")
    print("CHUNK PREVIEW")
    print(f"{'='*70}")

    for i, chunk in enumerate(chunks[:num_preview]):
        print(f"\n--- Chunk {i+1} ---")
        print(f"ID: {chunk['chunk_id']}")
        print(f"Size: {chunk['token_count']} tokens / {chunk['char_count']} chars")
        print(f"Text preview:")
        print(chunk['text'][:300] + "..." if len(chunk['text']) > 300 else chunk['text'])

        if chunk['previous_text']:
            print(f"\n📍 Previous context: ...{chunk['previous_text'][-100:]}")
        if chunk['next_text']:
            print(f"📍 Next context: {chunk['next_text'][:100]}...")

if __name__ == "__main__":
    all_chunks = chunk_folder(
         input_folder="/content/drive/MyDrive/data_sample/markdown_file",
         output_folder="/content/drive/MyDrive/data_sample/chunked_file",
         chunk_size=1000,
         chunk_overlap=200
     )
````

````python
import json
import os
import time
from typing import List, Dict, Tuple
from google.colab import drive
from pinecone import Pinecone, ServerlessSpec
from sentence_transformers import SentenceTransformer
from tqdm.auto import tqdm
````

````python
class vectorEmbedding:
  def __init__(self, pinecone_api_key: str, index_name: str, model_name: str = 'paraphrase-multilingual-mpnet-base-v2'):
    self.index_name = index_name
    self.model = SentenceTransformer(model_name)
    self.dimension = self.model.get_sentence_embedding_dimension()

    self.pc = Pinecone(api_key=pinecone_api_key)

    self._setup_index()


  def _setup_index(self):
    # Create an index if it does not exist
    if self.index_name not in self.pc.list_indexes().names():
        print(f"Creating Pinecone index '{self.index_name}'...")
        self.pc.create_index(
            name=self.index_name,
            dimension=self.dimension,
            metric='cosine',
            spec=ServerlessSpec(cloud='aws', region='us-east-1')
        )
        print(f"Index '{self.index_name}' created.")
    else:
        print(f"Index '{self.index_name}' already exists.")

    # Connect to the index
    self.index = self.pc.Index(self.index_name)
    print(f"Connected to Pinecone index '{self.index_name}'.")

  def load_data(self, json_path: str) -> List[Dict]:
        """
        Membaca JSON.
        - Jika formatnya Dictionary {"file": [chunks]}, diratakan jadi List [chunk, chunk].
        - Jika formatnya sudah List [chunk, chunk], langsung dikembalikan.
        """
        if not os.path.exists(json_path):
            raise FileNotFoundError(f"File tidak ditemukan: {json_path}")

        with open(json_path, 'r', encoding='utf-8') as f:
            raw_data = json.load(f)

        flat_data = []

        # JSON berbentuk Dictionary (File -> List of Chunks)
        if isinstance(raw_data, dict):
            print("📦 Terdeteksi struktur Dictionary. Meratakan data...")

            for filename, chunks in raw_data.items():
                if isinstance(chunks, list):
                    for chunk in chunks:
                        # Tambahkan nama file ke dalam chunk jika belum ada
                        if 'source_file' not in chunk:
                            chunk['source_file'] = filename
                        flat_data.append(chunk)
                else:
                    # Jika value-nya bukan list, anggap single object
                    flat_data.append(chunks)

        # JSON sudah berbentuk List Datar
        elif isinstance(raw_data, list):
            print("📦 Terdeteksi struktur List. Memproses langsung...")
            flat_data = raw_data

        else:
            raise ValueError("Format JSON tidak dikenali (harus List atau Dict)")

        print(f"📂 Siap memproses {len(flat_data)} chunks.")
        return flat_data

  def run_ingestion(self, json_path: str, text_key: str = "text", batch_size: int = 100):

        data = self.load_data(json_path)

        vectors_to_upsert = []
        print(f"🚀 Memulai proses embedding untuk {len(data)} chunks...")


        for i, item in tqdm(enumerate(data), total=len(data)):

            # Ambil teks
            text_content = item.get(text_key)

            # Validasi teks
            if not text_content or not isinstance(text_content, str):
                continue

            # Buat ID
            raw_id = item.get("chunk_id", f"chunk_{i}")
            chunk_id = str(raw_id).encode('ascii', 'ignore').decode('ascii')

            # Embedding
            embedding = self.model.encode(text_content).tolist()

            # Metadata
            metadata = {
                "text": text_content,
                "source": item.get("source_file", "unknown"),
                "title": item.get("document_title", "unknown"),
                "page": item.get("chunk_index", 0)
            }

            vectors_to_upsert.append((chunk_id, embedding, metadata))

            # Upload Batch
            if len(vectors_to_upsert) >= batch_size:
                self.index.upsert(vectors=vectors_to_upsert)
                vectors_to_upsert = []

        # Upload sisa
        if vectors_to_upsert:
            self.index.upsert(vectors=vectors_to_upsert)
        print("✨ Ingestion Selesai!")


  def search_with_timing(self, query: str, top_k: int = 3) -> Tuple[List, float]:
        start_time = time.perf_counter()
        query_vector = self.model.encode(query).tolist()
        results = self.index.query(vector=query_vector, top_k=top_k, include_metadata=True)
        duration = time.perf_counter() - start_time
        return results, duration
````

````python
import time
from groq import Groq

class RAGChatbot:
    def __init__(self, vector_pipeline, groq_api_key: str, llm_model: str = "llama3-8b-8192"):

        self.pipeline = vector_pipeline
        self.client = Groq(api_key=groq_api_key)
        self.model = llm_model

    def generate_response(self, query: str, top_k: int = 3):
        start_time = time.perf_counter()


        search_results, _ = self.pipeline.search_with_timing(query, top_k=top_k) #find relevant document in pinecone

        # konteks untuk LLM
        contexts = []
        sources = set() # Pakai set biarr tdk duplikat

        for match in search_results['matches']:
            text = match['metadata']['text']
            source = match['metadata'].get('source', 'Unknown')
            score = match['score']

            # threshold
            if score > 0.3:
                contexts.append(f"Isi Dokumen (Sumber: {source}):\n{text}")
                sources.add(source)

        context_str = "\n\n---\n\n".join(contexts)


        if not context_str:
            return {
                "answer": "Maaf, saya tidak menemukan informasi yang relevan dalam dokumen Anda untuk menjawab pertanyaan tersebut.",
                "sources": [],
                "time": time.perf_counter() - start_time,
                "context_used": ""
            }

        # Instruction
        system_prompt = """
        Kamu adalah asisten AI yang cerdas dan membantu.
        Tugasmu adalah menjawab pertanyaan pengguna BERDASARKAN konteks dokumen yang diberikan di bawah ini.

        Aturan:
        1. Jawablah dengan Bahasa Indonesia yang formal, jelas, dan lengkap.
        2. JANGAN mengarang jawaban. Jika jawaban tidak ada di konteks, katakan "Informasi tidak ditemukan di dokumen".
        3. Fokus pada fakta yang ada di teks.
        """

        user_message = f"""
        Konteks Dokumen:
        {context_str}

        Pertanyaan Pengguna:
        {query}
        """

        # generation
        chat_completion = self.client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_message}
            ],
            model=self.model,
            temperature=0.1, # avoid halusinasi
        )

        answer = chat_completion.choices[0].message.content


        end_time = time.perf_counter()

        return {
            "answer": answer,
            "sources": list(sources),
            "time": end_time - start_time,
            "context_used": context_str # Opsional: untuk debugging
        }
````
